```python
import pandas as pd
```

```python
df_e09 = pd.read_csv('Data/SingleEpisodes/rutweetE09.csv')
df_e10 = pd.read_csv('Data/SingleEpisodes/rutweetE10.csv')
df_e11 = pd.read_csv('Data/SingleEpisodes/rutweetE11.csv')
df_e12 = pd.read_csv('Data/SingleEpisodes/rutweetE12.csv')
df_e13 = pd.read_csv('Data/SingleEpisodes/rutweetE13.csv')
df_e16 = pd.read_csv('Data/SingleEpisodes/rutweetE16.csv')
```

```python
df_all = df_e09.append(df_e10).append(df_e11).append(df_e12).append(df_e13).append(df_e16)
df_all
```

```python
df_all['weight'] = df_all.groupby(['source', 'target'])['weight'].transform('sum')
df_all_weight = df_all.drop_duplicates(subset=['source', 'target'])
```

```python
df_all_weight
```

```python
df_all_weight.to_csv('Data/rutweet.csv',index=False)
```
