```python
rules: list = [
    {"value": hashtag+' has:mentions', "tag": "hashtagRu"},
    {"value": hashtag2+' has:mentions', "tag": "hashtag"},
] + [{'value': drag, 'tag': drag} for drag in drags]
    
stream = FilteredStream()
stream._endpoint+="?tweet.fields=author_id,entities,created_at"



print(stream.get_rules())

del_rules = {
            "delete": {
                "ids": ['1380647816729612298'] # example id
            }
}

#stream.delete_rule(del_rules)
stream.add_rule(data={"add": rules})
```

CREA UN FILE JSON CON DENTRO SCRITTO [] DI NOME FILEJSON

```python
filejson = 'rutweetE16.json'
```

```python
def collect_tweets(filejson):

    def collect_tweets_agent():

        def save_tweets(tweets,indent=1):
            print("Salvataggio su json")
            with open(filejson,'w') as fwrite:
                json.dump(tweets,fwrite,indent=indent)

        tweets = []
        print("Lettura dati già archiviati")
        with open(filejson,'r') as rfile:
                tweets = json.load(rfile)
        try:
            print("Acquisizione nuovi dati")
            for tweet in stream.connect():
                tweets.append(tweet)
                if len(tweets) > 1000000:
                    raise TypeError()

        except KeyboardInterrupt:
            save_tweets(tweets)
            return 'end'

        except requests.exceptions.ConnectionError:
            save_tweets(tweets)
            return 'connection error'

        except socket.timeout:
            save_tweets(tweets)
            return 'download error'

        except TypeError:
            save_tweets(tweets)
            return 'toolong'
        
        except ValueError:
            save_tweets(tweets)
            return 'decode error'
        
        except:
            save_tweets(tweets)
            return 'unexpected error'

        save_tweets(tweets)
        return 'closed'
    
    fail_count = 0
    n_file = 1
    while True:
        r = collect_tweets_agent()
        print(r)
        if r == 'end' : break
        if r == 'toolong':
            filejson = filejson.split('.')[0] + "_" + str(n_file) + '.json'
            n_file+=1
        if r == 'connection error' or r == 'download error': fail_count+=1
        time.sleep(2 ** fail_count)
```

```python
collect_tweets(filejson)
```
